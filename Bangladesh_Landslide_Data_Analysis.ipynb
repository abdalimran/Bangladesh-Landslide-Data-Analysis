# Data Analysis of Landslides in Bangladesh

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Bangladesh_Landslide_Cleaned_Data.csv')
data
```

# 1. Year vs Number of Landslides

```python
year_occurence = data['year'].value_counts()
year_occurence = year_occurence.sort_index()

years = year_occurence.index
y_occurrences = year_occurence.values
```

```python
y_pos = np.arange(len(years))
 
plt.bar(y_pos, y_occurrences, align='center')
plt.xticks(y_pos, years)
plt.xlabel('Years')
plt.ylabel('Number of Landslides')
plt.title('Year vs Number of Landslides')
plt.tight_layout()

plt.rcParams['axes.facecolor']='#e0f2f1'
plt.rcParams['savefig.facecolor']='#e0f2f1'

plt.savefig('01_year_vs_nlandslides.jpg', dpi = 180)
plt.show()
```

# 2. Month vs Number of Landslides

```python
month_occurence = data['month_number'].value_counts()
month_occurence = month_occurence.sort_index()

month_label = ['','January','February','March','April','May','June','July','August','September','October','November','December']

months =  []
m_occurrences = []

for i in month_occurence.index:
    months.append(month_label[i])
    m_occurrences.append(month_occurence[i])
```

```python
y_pos = np.arange(len(months))
 
plt.barh(y_pos, m_occurrences, align='center', color='darkslategrey')
plt.yticks(y_pos, months)
plt.ylabel('Months')
plt.xlabel('Number of Landslides')
plt.title('Month vs Number of Landslides')
plt.tight_layout()

plt.rcParams['axes.facecolor']='#e0f2f1'
plt.rcParams['savefig.facecolor']='#e0f2f1'

plt.savefig('02_month_vs_nlandslides.jpg', dpi = 1000)
plt.show()
```

# 3. City vs Number of Landslides

```python
city_occurence = data['city'].value_counts()
city_occurence = city_occurence.sort_index()

cities = city_occurence.index
c_occurrences = city_occurence.values
```

```python
x = list(range(0,110,5))
y = city_occurence

plt.plot(x, y, marker='o',color='navy',linestyle='-')
plt.xticks(x, cities, rotation='vertical')
plt.ylim(0, 8)
plt.xlabel('Cities')
plt.ylabel('Number of Landslides')
plt.title('City vs Number of Landslides')
plt.tight_layout()

plt.rcParams['axes.facecolor']='#e0f2f1'
plt.rcParams['savefig.facecolor']='#e0f2f1'

plt.savefig('03_city_vs_nlandslides.jpg', dpi = 1000)
plt.show()
```

# 4. Landslide Trigger vs Landslide Type

```python
landslide_type = list(data['landslide_type'])
landslide_trigger = list(data['trigger'])

landslide_type_trigger = pd.DataFrame({'landslide_type': landslide_type, 'trigger': landslide_trigger})
landslide_type_trigger = landslide_type_trigger.sort_values(by=['landslide_type'], ascending=True)

landslide_df = landslide_type_trigger.iloc[:29]
mudslide_df = landslide_type_trigger.iloc[29:]

landslide_trigger_occ = landslide_df['trigger'].value_counts()
mudslide_trigger_occ = mudslide_df['trigger'].value_counts()
```

```python
plt.axis("equal")
plt.pie(
    list(landslide_df['trigger'].value_counts().values),
    labels=list(landslide_df['trigger'].unique()),
    autopct="%1.1f%%",
    explode = (0.05, 0, 0),
    startangle=90
)
plt.title("Trigger percentage of Landslide")

plt.rcParams['axes.facecolor']='#e0f2f1'
plt.rcParams['savefig.facecolor']='#e0f2f1'

plt.tight_layout()
plt.savefig('04_landslide_type_vs_trigger.jpg', dpi = 1000)

plt.show()
```

```python
plt.axis("equal")
plt.pie(
    list(mudslide_df['trigger'].value_counts().values),
    labels=list(mudslide_df['trigger'].unique()),
    autopct="%1.1f%%",
    explode = (0.05, 0),
    colors= ('tomato','deepskyblue'),
    startangle=90
    )
plt.title("Trigger percentage of Mudslide")

plt.rcParams['axes.facecolor']='#e0f2f1'
plt.rcParams['savefig.facecolor']='#e0f2f1'

plt.tight_layout()
plt.savefig('05_mudslide_type_vs_trigger.jpg', dpi = 1000)

plt.show()
```

# 5. Landslide Types vs Fatalilties

```python
years = list(data['year'])
types = list(data['landslide_type'])
fatalities = list(data['fatalities'])

ytf_df = pd.DataFrame({"years":years,"types":types,"fatalities":fatalities})

ytf_df = ytf_df.sort_values(by=['types'], ascending=True)

landslide_ytf_df = ytf_df.iloc[:29]
mudslide_ytf_df = ytf_df.iloc[29:]

mudslide_fat_yr = []
for y in data['year'].unique():
    mudslide_fat_yr.append(mudslide_ytf_df.loc[mudslide_ytf_df['years'] == y, 'fatalities'].sum())

landslide_fat_yr = []
for y in data['year'].unique():
    landslide_fat_yr.append(landslide_ytf_df.loc[landslide_ytf_df['years'] == y, 'fatalities'].sum())
    
```

```python
N = len(data['year'].unique())

ind = np.arange(N)
width = 0.35 

ax = plt.subplot()
land_fat_yr = ax.bar(ind, landslide_fat_yr, width, color='tomato')
mud_fat_yr = ax.bar(ind+width, mudslide_fat_yr, width, color='navy')

ax.set_ylabel('Number of Fatalities')
ax.set_xlabel('Year')
ax.set_title('Fatalities Every Year by Landslide Types')
ax.set_xticks(ind + width / 2)
ax.set_yticks(np.arange(min(mudslide_fat_yr), max(landslide_fat_yr)+1, 10))
ax.set_xticklabels(data['year'].unique())
ax.legend((land_fat_yr[0], mud_fat_yr[0]), ('Landslides', 'Mudslides'))

plt.tight_layout()
plt.savefig('06_fatalities_yr_types.jpg', dpi = 1000)

plt.show()
```
